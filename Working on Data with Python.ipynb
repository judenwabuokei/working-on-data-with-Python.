## Filtering and Ordering

```python
import pandas as pd

# to ignore warnings
import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv('world_population.csv')
```

```python
df.head()
```

#### Renaming the Country/Territory column

```python
df.columns = df.columns.str.replace('Country/Territory','Country')
df
```

####  Top 10 countries with the highest population

```python
df[df.Rank <=10]
```

#### Filtering out specific countries from the dataset


```python
selected_countries = ['Ghana', 'Morocco', 'Egypt', 'Nigeria']
df[df.Country.isin(selected_countries )]
```

####  Filtring out countries with certain strings in their name

```python
df[df.Country.str.contains('United')]
```

 #### Setting the  index as the country column

```python
df1 = df.set_index('Country')
df1
```

 #### Filtering out Columns from dataset

```python
df1.filter(items=['Continent','Rank', '2022 Population'])
```

####  Filtering out Country names

```python
df1.filter(items=['Algeria','Nigeria', 'Libya'], axis =0)
```

#### Filtering out countries name like --

```python
df1.filter(like='ria', axis =0)
```

#### Getting a Country details from the dataset

```python
df1.loc['Ghana']
```

### Ordering in pandas

#### Ordering by column 

```python
df[df.Rank <=10].sort_values(by='1970 Population', ascending=False)
```

#### Ordering by a list

```python
df[df.Rank <=10].sort_values(by=['Country','Continent'], ascending=True)
```

#### Ordering by a list and mapped ascending options

```python
df[df.Rank <=10].sort_values(by=['Country','Continent'], ascending=[False, True])
```

### Indexing

```python
df1 = df.set_index(['Continent','Country'])                      # use df1.reset_index() to reset the index 
df1
```

```python
df1.sort_index(ascending=[True,False])
```

### Group by and Aggregating 

```python
df.groupby('Continent').mean()
```

#### Group by and Aggregate

```python
df.groupby('Continent').agg({'2022 Population':['mean'], '2020 Population':['mean'], '2015 Population':['mean'], '2010 Population':['mean'], })
```

#### Group by multiple columns

```python
df.groupby(['Continent','']).mean()
```

### Data Cleaning

```python
customers = pd.read_excel('Customer Call List.xlsx')
```

```python
customers
```

#### Renaming a Column

```python
 customers.columns = customers.columns.str.replace('Paying Customer','Paying_Customer')
```

```python
customers
```

####  Drop Duplicates

```python
customers = customers.drop_duplicates()
```

```python
customers
```

#### Droping Columns

```python
customers = customers.drop(columns='Not_Useful_Column') 
```

```python
customers
```

#### Cleaning Last_Name Column

```python
customers.Last_Name = customers.Last_Name.str.strip('/._')          #strip removes unwanted char thats left or right of the data
```

```python
customers
```

#### Cleaning Phone_Number	 Column

```python
customers.Phone_Number = customers.Phone_Number.str.replace('[^a-zA-Z0-9]','')
```

```python
customers
```

```python
customers.Phone_Number = customers.Phone_Number.apply(lambda x: str(x))   # Converts the column to string
```

```python
customers.Phone_Number = customers.Phone_Number.apply(lambda x: x[0:3] + '-' + x[3:6] + '-' +x[6:])

# This code converts the phone numbers into the format xxx-xxx-xxxx
```

```python
customers
```

```python
customers.Phone_Number = customers.Phone_Number.str.replace('nan--','')
customers.Phone_Number = customers.Phone_Number.str.replace('Na--','')
```

```python
customers
```

#### Cleaning Address Column

```python
 customers[['Street_Address','State','Zip_Code']] = customers.Address.str.split(',',2, expand = True)
```

```python
customers
```

#### Cleaning Paying_Customer Column

```python
customers.Paying_Customer = customers.Paying_Customer.str.strip('oes')
```

```python
customers
```

#### Cleaning Do_Not_Contact Column

```python
customers.Do_Not_Contact = customers.Do_Not_Contact.str.strip('oes')
```

```python
customers
```

 #### Dropping Address Column

```python
customers = customers.drop(columns='Address') 
```

```python
customers
```

```python
customers = customers.fillna('')
```

```python
customers
```
